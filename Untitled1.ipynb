```python
# Cell 1: Imports and Data Loading
import tensorflow as tf
from tensorflow.keras import layers, models
import gradio as gr
import numpy as np
from PIL import Image

# 1. Load the MNIST dataset built directly into Keras
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Preprocess the data
# Scale pixel values to be between 0.0 and 1.0
x_train, x_test = x_train / 255.0, x_test / 255.0

# Reshape data to add a channel dimension (grayscale = 1 channel)
# Expected shape: (num_images, 28, 28, 1)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

print(f"Data loaded successfully! Training shape: {x_train.shape}")
```

```python
# Cell 2: Model Architecture and Training

# 1. Define the Sequential CNN model
model = models.Sequential([
    # First Convolutional Layer: extracts basic edge features
    layers.Conv2D(16, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(pool_size=(2, 2)),

    # Second Convolutional Layer: extracts more complex shape combinations
    layers.Conv2D(32, kernel_size=(5, 5), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),

    # Flattening the 2D matrices into a 1D vector for classification
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax') # Softwalk outputs probabilities for digits 0-9
])

# 2. Compile the model specifying the math metrics
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', # Used when labels are integers (0-9)
    metrics=['accuracy']
)

# 3. Train the model for 3 epochs
print("Training started...")
model.fit(x_train, y_train, epochs=3, batch_size=64, validation_data=(x_test, y_test))
print("Training finished!")
```

```python
# Cell 3: Live Canvas App Interface (UPDATED)
import numpy as np
import gradio as gr
from PIL import Image, ImageOps

def predict_digit(sketchpad_image):
    if sketchpad_image is None:
        return "Draw a digit!"

    # Extract the image from Gradio's canvas payload
    img = sketchpad_image["composite"] if isinstance(sketchpad_image, dict) else sketchpad_image

    # 1. Convert to RGB to safely drop any weird web transparencies
    img = img.convert("RGB")

    # 2. Convert to Grayscale
    img = img.convert("L")

    # 3. THE FIX: Invert the colors! (White background becomes black)
    img = ImageOps.invert(img)

    # 4. Resize to 28x28
    img = img.resize((28, 28))

    # Convert to numpy array and normalize [0, 255] to [0.0, 1.0]
    img_array = np.array(img) / 255.0

    # Add dimensions: (1, 28, 28, 1)
    img_array = np.expand_dims(img_array, axis=(0, -1))

    # Run prediction
    predictions = model.predict(img_array, verbose=0)[0]

    return {str(i): float(predictions[i]) for i in range(10)}

# Configure the Gradio layout interface
interface = gr.Interface(
    fn=predict_digit,
    inputs=gr.Sketchpad(type="pil", label="Draw a single digit (0-9)"),
    outputs=gr.Label(num_top_classes=3, label="Predictions"),
    title="TensorFlow Real-Time Digit Classifier"
)

interface.launch(share=True)
```

```python
# Cell 4: Plot Loss and Accuracy Curves
import matplotlib.pyplot as plt

# Re-run training but capture the history this time to ensure we have the data
print("Training a fresh model to capture history statistics...")
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

# Create a figure with two subplots side-by-side
plt.figure(figsize=(14, 5))

# Subplot 1: Accuracy Curve
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='x')
plt.title('Model Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Subplot 2: Loss Curve
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='x')
plt.title('Model Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Cell 5: Generate and Plot Confusion Matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 1. Get predictions for the entire test dataset
print("Generating predictions for the 10,000 test cases...")
test_predictions = model.predict(x_test, verbose=0)
predicted_labels = np.argmax(test_predictions, axis=1)

# 2. Compute the confusion matrix
cm = confusion_matrix(y_test, predicted_labels)

# 3. Plot using Seaborn's heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=[str(i) for i in range(10)],
            yticklabels=[str(i) for i in range(10)])

plt.title('Test Dataset Confusion Matrix')
plt.xlabel('Predicted Digit Label')
plt.ylabel('True Digit Label')
plt.show()

# 4. Print overall final test metrics
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nFinal Test Set Results:")
print(f"-> Test Loss: {test_loss:.4f}")
print(f"-> Test Accuracy: {test_acc * 100:.2f}%")
```

### Examples of Misclassified Images
